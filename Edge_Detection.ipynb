1️⃣ Lane Detection in Self-Driving Cars


```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload your image
uploaded = files.upload()
```

```python

# Step 2: Get the filename (for example, tesla.jpg)
img = cv2.imread("tesla.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Optional: verify it's loaded
if img is None:
    raise FileNotFoundError("Image not found! Make sure the file name matches exactly.")
```

```python
# Step 3: Apply edge detection
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobelx, sobely)
laplacian = cv2.Laplacian(img, cv2.CV_64F)
canny = cv2.Canny(img, 80, 150)
```

```python
# Step 4: Show results
titles = ['Original', 'Sobel', 'Laplacian', 'Canny']
images = [img, sobel, laplacian, canny]

plt.figure(figsize=(12,4))
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
